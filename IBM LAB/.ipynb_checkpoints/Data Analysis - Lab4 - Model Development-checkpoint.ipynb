In this lab, we will create some models which will predict the price of the car using the variables or features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

---
### Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
x = df[['highway-mpg','normalized-losses','horsepower']]
y = df['price']
lm = LinearRegression()
lm.fit(x,y)
```

```python
yhat = lm.predict(x)
print('Intercept: ',lm.intercept_)
print('Coef: ', lm.coef_)
print('R-squared: ', lm.score(x,y))
```

### Piplines

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
Input = [('scale', StandardScaler()), ('polynomial', PolynomialFeatures(degree = 2)),('mode', LinearRegression())]
```

```python
pipe = Pipeline(Input)
pipe.fit(z,y)
yhat = pipe.predict(z)
```

### Cross validation
when the data is not large enough to be split into train and test sets, we use cross_val_score

```python
from sklearn.model_selection import cross_val_score
```

```python
#Define the variables
x = df.drop('price', axis = 1)
y = df['price']
x
```

```python
#Build Model
lr1 = LinearRegression()
lr1.fit(x[['horsepower']],y)
CVS = cross_val_score(lr1, x[['horsepower']],y, cv=4)
```

```python
#Calculate the R-square, the mean of the folds
print ('R-Squared = ',(lr1.score(x[['horsepower']],y)))
print ('Mean of folds = ', CVS.mean())
```

```python
CVS
```

### Use the <font color = 'red'>"Cross_val_predict"</font>

```python
from sklearn.model_selection import cross_val_predict
yhat = cross_val_predict(lr1, x[['horsepower']], y, cv = 4)
yhat[:4]
```

## Multiple features with polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
x_data = df[['horsepower', 'curb-weight','engine-size','highway-mpg']]
y_data = df[['price']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state = 16)
print ('x_train size:', x_train.shape)
print('x_test size: ', x_test.shape)
```

```python
pr = PolynomialFeatures(degree = 2, include_bias = False)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
print ('Dimension of x_train_pr: ', x_train_pr.shape)
print('Dimension of x_test_pr: ', x_test_pr.shape)
```

```python
pr = LinearRegression()
pr.fit(x_train_pr, y_train)
```

```python
yhat = pr.predict(x_train_pr)
y_hat_test = pr.predict(x_test_pr)
```

```python
plt.figure(figsize = (12,10))
ax1 = sns.distplot(y_test,hist = False, color = 'b', label = 'Actual Test dataset')
sns.distplot(y_hat_test, hist = False, color = 'r', label = 'Predicted Test dataset', ax = ax1)
plt.xlabel('Price',)
plt.ylabel('Proportion of cars')
plt.title ('Distribution of Actual Target Values and Predicted Target Values')
plt.show()
plt.close()
```

```python
# plot the residual plot
ax2 = sns.residplot(y_hat_test, y_test,color = 'r' )
sns.residplot(yhat, y_train, color = 'b', ax = ax2)
plt.show()
```

## Ridge Regression

Use alpha to prevent overfiiting

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pr = PolynomialFeatures(degree = 2, include_bias = False)
x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
RidgeModel = Ridge(alpha = 1)
RidgeModel.fit(x_train_pr,y_train)
print('R^2 values of training dataset: ',RidgeModel.score(x_train_pr,y_train ))
print ('R^2 values of testing dataset:', RidgeModel.score(x_test_pr, y_test))
```

```python
yhat_Ridge = RidgeModel.predict(x_test_pr)
```

```python
plt.figure(figsize = (12,10))
ax1 = sns.distplot(y_test,hist = False, color = 'b', label = 'Actual Test dataset')
sns.distplot(yhat_Ridge, hist = False, color = 'r', label = 'Predicted - Ridge', ax = ax1)
sns.distplot(y_hat_test, hist = False, color = 'y', label = 'Predicted - Linear', ax = ax1)
plt.xlabel('Price',)
plt.ylabel('Proportion of cars')
plt.title ('Distribution of Actual Target Values and Predicted Target Values')
plt.show()
plt.close()
```

## Grid Search

Allows us to scan through the free hyperparameter (ex. alpha in Ridge Model)

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyperpara = [{'alpha': [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000, 1000000]}]
Ridge_Grid = Ridge()
GridSearch1 = GridSearchCV(Ridge_Grid, hyperpara, cv =3)
GridSearch1.fit(x_train_pr, y_train)
```

```python
best_alpha = GridSearch1.best_estimator_

#Find the R^2 value of the best estimator
best_alpha.score(x_test_pr, y_test)
```
