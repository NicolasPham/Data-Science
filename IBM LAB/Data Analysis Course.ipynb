```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',26)
```

```python
# Import data
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(url, names = headers)
print('Data has been loaded')
```

```python
df
```

```python
# Replace "?" into "NaN"
df.replace('?', np.nan, inplace = True)
miss_df = pd.DataFrame()
miss_df['Miss value count'] = df.isnull().sum()
miss_df['Percentage'] = (100*df.isnull().sum()/len(df)).round(1)
miss_df
```

```python
# Since the maximum percentage of missing value is 20%, therefore we can drop all the missing value
df.dropna(inplace = True)
df
```
